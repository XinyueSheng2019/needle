```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os

all_transients_path = '/Users/xinyuesheng/Documents/astro_projects/data/ztf_transients_all.csv'
slsn_data_path = '/Users/xinyuesheng/Documents/astro_projects/scripts/classifier_v1/plotting/outputs/old_slsn_z_mag.json'
host_path = '/Users/xinyuesheng/Documents/astro_projects/data/host_info_r5_ext/'
transient_mags_path = '/Users/xinyuesheng/Documents/astro_projects/data/mag_sets_v4/'
```

```python
# 1. peak apparent magnitude distributions of the SLSNe with and without hosts (g and r bands)
# transient_z = pd.read_csv(transient_z_path)
import json
with open(slsn_data_path, 'r') as f:
    slsn_data = json.load(f)

all_transients = pd.read_csv(all_transients_path)
slsn_objs = all_transients[all_transients['type'] == 'SLSN-I']['ZTFID'].values

hosted_data = []
hostless_data = []
for obj in slsn_objs:
    obj_host_path = os.path.join(host_path, f'{obj}.csv')
    try:
        obj_data = [float(slsn_data[obj]['z']), float(slsn_data[obj]['mag'])]
        if os.path.exists(obj_host_path):
        
            hosted_data.append(obj_data)
        else:
            if obj_data[1] < 17:
                print(obj)
            hostless_data.append(obj_data)
    except:
        print(f'{obj} has no data')

hosted_data = np.array(hosted_data)
hostless_data = np.array(hostless_data)
```

```python

plt.figure(figsize=(10, 5))
plt.hist(hosted_data[:, 1], bins=30, alpha=0.5, label='SLSN-I with hosts')
plt.hist(hostless_data[:, 1], bins=30, alpha=0.5, label='SLSN-I without hosts')
plt.xlabel('Peak apparent magnitude (g band)')
plt.ylabel('Number of transients')
plt.legend()
plt.show()
```
